```python
from sklearn.linear_model import LinearRegression
```

```python
# Prepare the data
X = df['hours_studied'].values.reshape(-1, 1) # Independent variable (hours studied)
y = df['pass_exam'].values  # Dependent variable (pass exam)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Visualize the data and the regression line
plt.scatter(df['hours_studied'], df['pass_exam'], color='blue', label='Data points')  # Scatter plot of original data
plt.plot(df['hours_studied'], y_pred, color='red', label='Regression line')  # Fitted regression line

# Add labels and title
plt.title("Linear Regression: Hours Studied vs Pass Exam")
plt.xlabel("Hours Studied")
plt.ylabel("Pass Exam (0=Fail, 1=Pass)")

# Show the plot
plt.legend()
plt.show()

# Print the regression coefficients
print(f"Regression Coefficient (Slope): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
# Convert columns to numeric if needed
df['exam_score'] = pd.to_numeric(df['exam_score'], errors='coerce')
df['hours_studied'] = pd.to_numeric(df['hours_studied'], errors='coerce')
df['attendance'] = pd.to_numeric(df['attendance'], errors='coerce')
df['tutoring_sessions'] = pd.to_numeric(df['tutoring_sessions'], errors='coerce')
```

```python
x = df[['hours_studied', 'attendance','tutoring_sessions']]
y = df['exam_score']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=15)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
LR.fit(x_train.values,y_train)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
print("The LR model is: Y = ",LR.intercept_, "+", LR.coef_[0], "hours_studied + ", LR.coef_[1], "attendance+ ", LR.coef_[2], "tutoring_sessions")
```

```python
student_score = [[30, 70, 5]]

pred = LR.predict(student_score)
print("prediction= ", pred[0])
```

```python
y_pred = LR.predict(x_test.values)
y_pred
```

```python
sns.regplot(x = y_test, y = y_pred, color='green')
```

```python
from sklearn import metrics
import numpy as np

mae = metrics.mean_absolute_error(y_test,y_pred)
mse = metrics.mean_squared_error(y_test,y_pred)
rmse = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
print('mae =',mae.round(2))
print('mse =',mse.round(2))
print('rmse =',rmse.round(2))
```
